```kotlin
import java.io.File
val words = File("words.txt").readText(Charsets.UTF_8)
    .split(",")
    .map { it.trim().replace("\"","") }
val answers = File("answers.txt").readText(Charsets.UTF_8)
    .split(",")
    .map { it.trim().replace("\"","") }
print(words)
```

```kotlin
val freqs = HashMap<Char, Int>()

words.forEach { word ->
    word.forEach { char ->
        freqs.put(char, freqs.getOrDefault(char, 0) + 1)
    }
}
println(freqs)
```

```kotlin
println(freqs.toSortedMap(compareByDescending {freqs[it]}))
```

```kotlin
val scores = HashMap<String, Int>()

words.forEach { word ->
    val usedLetters = HashSet<Char>()
    word.forEach { char ->
        if(!usedLetters.contains(char)){
            scores.put(word, scores.getOrDefault(word, 0) + freqs[char]!!)
            usedLetters.add(char)
        }
    }
}
println(scores.toSortedMap(compareByDescending {scores[it]}))
```

```kotlin
val freqsP = HashMap<Pair<Char, Int>, Int>()

words.forEach { word ->
    word.forEachIndexed { index, char ->
        freqsP.put(char to index, freqsP.getOrDefault(char to index, 0) + 1)
    }
}
println(freqsP.toSortedMap(compareByDescending {freqsP[it]}))
```

```kotlin
val scoresP = HashMap<String, Int>()

words.forEach { word ->
    val usedLetters = HashSet<Char>()
    word.forEachIndexed { index, char ->
        if(!usedLetters.contains(char)){
            scoresP.put(word, scoresP.getOrDefault(word, 0) + freqsP[char to index]!!)
            usedLetters.add(char)
        }
    }
}
println(scoresP.toSortedMap(compareByDescending {scoresP[it]}))
```

```kotlin
val freqsU = HashMap<Char, Int>()

words.forEach { word ->
    word.toSet().joinToString("").forEach { char ->
        freqsU.put(char, freqsU.getOrDefault(char, 0) + 1)
    }
}
println(freqsU)
```

```kotlin
val scoresU = HashMap<String, Int>()

words.forEach { word ->
    val usedLetters = HashSet<Char>()
    word.forEach { char ->
        if(!usedLetters.contains(char)){
            scoresU.put(word, scoresU.getOrDefault(word, 0) + freqsU[char]!!)
            usedLetters.add(char)
        }
    }
}
println(scoresU.toSortedMap(compareByDescending {scoresU[it]}))
```

```kotlin
val scoresPWeighted = HashMap<String, Int>()
val weight = 1

words.forEach { word ->
    val usedLetters = HashSet<Char>()
    word.forEachIndexed { index, char ->
        if(!usedLetters.contains(char)){
            scoresPWeighted.put(word, scoresPWeighted.getOrDefault(word, 0) + (freqsP[char to index]!!*weight).toInt() + freqs[char]!!)
            usedLetters.add(char)
        }
    }
}
println(scoresPWeighted.toSortedMap(compareByDescending {scoresPWeighted[it]}))
```
